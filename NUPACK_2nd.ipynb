```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#strandA= []max
#strand B = 1
ratio_value = [0.08, 0.2,  0.8, 1, 1.3, 2.5, 5]
y_value_complex = [30/970, 49/951, 0.1, 1/9, 300/1000, 1.5/0.99, 4]
plt.plot(ratio_value, y_value_complex,marker= 'o' )
plt.axvline (1, linestyle= "dashed", color ="red")
plt.xlabel ("[strand C] mM")
plt.ylabel ("[AC]/ [AB]")
plt.title("Concentration - Complexes Analysis ")
plt.show()
```

```python
T= [10, 20, 30, 40 , 50 , 60 , 70, 90]
ac= [100, 120, 150, 180, 210, 250, 290, 320]
sC = [222, 420, 560, 610, 620, 630, 600, 590]
ab= [900, 880, 850, 820, 780, 750, 700, 560]
plt.plot(T, ac, marker='o', label= '[AC]', markersize= 4)
plt.plot(T, sC, color="green", label= "[single strand C ] / [AB]", linestyle= "dashed", marker='o', markersize= 4)
plt.plot(T, ab, label= "[AB]", marker='o', markersize= 4)
plt.xlabel(" T°C")
plt.ylabel(" [ ] mM ")
plt.legend()
plt.title ("Complexes Ratio over T°C")
```
